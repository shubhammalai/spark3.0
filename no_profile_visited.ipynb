```python
from pyspark.sql import SparkSession
import getpass

username = getpass.getuser()

spark = SparkSession.builder \
    .appName("MySparkApp") \
    .master("yarn") \
    .config("spark.ui.port", "0") \
    .config("spark.sql.warehouse.dir", f"/user/{username}/warehouse") \
    .enableHiveSupport() \
    .getOrCreate()
```

```python
spark
```

```python
base_rdd  = spark.sparkContext.textFile("/user/itv023333/data/input/linkedin_views.csv")
```

```python
base_rdd.collect()
```

```python
mapped_rdd = base_rdd.map(lambda x:(x.split(",")[2],1))
```

```python
mapped_rdd.collect()
```

```python
reduced_rdd = mapped_rdd.reduceByKey(lambda x,y:x+y)
```

```python
reduced_rdd.collect()
```

```python
reduced_rdd.saveAsTextFile("/user/itv023333/data/linkedin_output")
```

```python
spark.stop()
```
