Update column name order_status to status in **df_json** df

```python
df_json2 = df_json.withColumnRenamed("order_status","status")
df_json2.show()
```

change datatype from string to timestamp in df_json df

```python
from pyspark.sql.functions import *
df_json3 = df_json2.withColumn("order_date_new",to_timestamp(df_json2.order_date))
df_json4 = df_json3.drop("order_date")
df_json4.show()
```

```python
df_json4.printSchema()
```

```python
df_json5 = df_json4.filter("status= 'CLOSED'")
# df_json5 = df_json4.where("status= 'CLOSED'")
df_json5.show()
df_json5.count()
```

read a table

```python
df_table  = spark.read.table("workspace.spark_practice.orders_table")
df_table.show(5)
df_table.count()
```

create table from df

```python
df_json5.createOrReplaceTempView("orders_table")
```

```python
df = spark.sql("select * from orders_table where status = 'CLOSED'")
df.show()
df.count()
```
