```python
from pyspark.sql import SparkSession
import getpass

username = getpass.getuser()

spark = SparkSession.builder \
    .appName("MySparkApp") \
    .master("yarn") \
    .config("spark.ui.port", "0") \
    .config("spark.sql.warehouse.dir", f"/user/{username}/warehouse") \
    .enableHiveSupport() \
    .getOrCreate()
```

```python
base_rdd = spark.sparkContext.textFile("data/input/inputfile.txt")
```

```python
base_rdd.take(5)
```

```python
trans_rdd = base_rdd.flatMap(lambda x:(x.split(" ")))
```

```python
trans_2 = trans_rdd.map(lambda x:(x,1))
```

```python
trans_3 = trans_2.reduceByKey(lambda x,y:x+y)
```

```python
trans_3.collect()
```
